```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
      if (self.age >= 20):
          print(f"{self.name} は大人です")
      else:
          print(f"{self.name} は大人ではありません")

hanako = Human("花子", 23)
taro = Human("太郎", 19)
jirou = Human("次郎", 28)

human_list = [hanako, taro, jirou]

for human in human_list:
    human.check_adult()
```
